```python
PPDB_SEPERATOR = "|||"
PPDB_SOURCE_LOCATION = 1
PPDB_TARGET_LOCATION = 2
def load_ppdb_lexical():
    lines = []
    with open("ppdb-1.0-s-lexical") as input_file:
        for line in input_file:
            lines.append(line)
    return lines

def get_equivalents_ppdb():
    ppdb_lines = load_ppdb_lexical()
    source_target = dict()
    for line in ppdb_lines:
        splitted = line.split(PPDB_SEPERATOR)
        source = splitted[PPDB_SOURCE_LOCATION]
        target = splitted[PPDB_TARGET_LOCATION]
        source = source.strip()
        target = target.strip()
        if not source in source_target:
            source_target[source]=[]
        source_target[source].append(target)
    return source_target

def get_word_equivalent_to(word,dictionary):
    if not word in source_target_ppdb:
        return []
    else:
        options = source_target_ppdb[word]
        return list(set(options))

source_target_ppdb = get_equivalents_ppdb()
print(get_word_equivalent_to("budgetary",source_target_ppdb))


        
```

```python
translation_memory = [line.lower().strip() for line in open('out.source')]
print(len(translation_memory))
query = 'Reference Guide for Budget Control'
query = query.lower()

def get_recursive_sentences(sentence_so_far,remaining_words):
    if len(remaining_words)==0:
        return [sentence_so_far]
    total = []
    for w in remaining_words[0]:
        s_so_far = sentence_so_far + " " + w
        from_here = get_recursive_sentences(s_so_far,remaining_words[1:])
        total.extend(from_here)
    return total

def get_alternative_sentences(query,dictionary):
    words = query.split()
    alternative_words = []
    for w in words:
        alternatives_w = get_word_equivalent_to(w,dictionary)
        alternatives_w.append(w)
        alternative_words.append(alternatives_w)
    alternatives = get_recursive_sentences("",alternative_words)
    # remove space:
    alternatives = [a[1:] for a in alternatives]
    return alternatives

def get_possibilities_from_tmx(query, dictionary, translation_memory):
    alternatives = get_alternative_sentences(query,dictionary)
    known = []
    for a in alternatives: 
        a = a.strip()
        if a in translation_memory:
            known.append(a)
    return known

total_replacable = 0
for query in translation_memory:
    alternatives = get_possibilities_from_tmx(query,source_target_ppdb,translation_memory)
    if len(alternatives) > 1:
        print(alternatives)
        total_replacable +=1
print(str(total_replacable//2) + " out of " + len(translation_memory))
        
```
